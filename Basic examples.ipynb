## Plot multiple curves in a single plot

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

# draw line with explicit x and y
plt.plot([1,2,4,8], [1,2,4,8], 'green')

# draw line with implicit y
plt.plot([1,2,4,8], 'blue')
plt.ylabel('some numbers')
plt.show()

```

## Plot multple plots using subplots

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(211) # numrows, numcols, fignum 
plt.plot(t1, f(t1), 'blue')

plt.subplot(212)
plt.plot(t1, f(t1), 'red')
plt.show()
```

## Plot histogram

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# x is a list of 10K floats

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)



plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$') # text inside the plot
plt.axis([40, 160, 0, 0.03]) # axis range, optional
plt.grid(True)
plt.show()
```
